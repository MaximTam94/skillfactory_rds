## 1. Цель и задачи проекта
#### Цель проекта - отследить влияние условий жизни учащихся в возрасте от 15 до 22 лет на их успеваемость по математике, чтобы на ранней стадии выявлять студентов, находящихся в группе риска.
#### Достигнуть цели поможет построенная модель, предсказывающая результаты госэкзамена по математике для каждого ученика школы. Для построения будущей модели должны быть выполнены следующие задачи:
#### - проведен разведывательный анализ данных;
#### - отобраны важные для модели признаки;
#### - составлен отчёт по результатам анализа.

## 2. Краткая информация о данных
#### Датасет содержит следующие данные:
#### 1. school - аббревиатура школы, в которой учится ученик. Номинативная переменная;
#### 2. sex - пол ученика ('F' - женский, 'M' - мужской). Номинативная переменная;
#### 3. age - возраст ученика (от 15 до 22). Количественная переменная;
#### 4. address - тип адреса ученика ('U' - городской, 'R' - за городом). Номинативная переменная;
#### 5. famsize - размер семьи('LE3' <= 3, 'GT3' >3). Номинативная переменная;
#### 6. Pstatus - статус совместного жилья родителей ('T' - живут вместе 'A' - раздельно). Номинативная переменная;
#### 7. Medu - образование матери (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее). Количественная переменная;
#### 8. Fedu - образование отца (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее). Количественная переменная;
#### 9. Mjob - работа матери ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое). Номинативная переменная;
#### 10. Fjob - работа отца ('teacher' - учитель, 'health' - сфера здравоохранения, 'services' - гос служба, 'at_home' - не работает, 'other' - другое). Номинативная переменная;
#### 11. reason - причина выбора школы ('home' - близость к дому, 'reputation' - репутация школы, 'course' - образовательная программа, 'other' - другое). Номинативная переменная;
#### 12. guardian - опекун ('mother' - мать, 'father' - отец, 'other' - другое). Номинативная переменная;
#### 13. traveltime - время в пути до школы (1 - <15 мин., 2 - 15-30 мин., 3 - 30-60 мин., 4 - >60 мин.). Количественная переменная;
#### 14. studytime - время на учёбу помимо школы в неделю (1 - <2 часов, 2 - 2-5 часов, 3 - 5-10 часов, 4 - >10 часов). Количественная переменная;
#### 15. failures - количество внеучебных неудач (n, если 1<=n<3, иначе 0). Количественная переменная;
#### 16. schoolsup - дополнительная образовательная поддержка (yes или no). Номинативная переменная;
#### 17. famsup - семейная образовательная поддержка (yes или no). Номинативная переменная;
#### 18. paid - дополнительные платные занятия по математике (yes или no). Номинативная переменная;
#### 19. activities - дополнительные внеучебные занятия (yes или no). Номинативная переменная;
#### 20. nursery - посещал детский сад (yes или no). Номинативная переменная;
#### 21. higher - хочет получить высшее образование (yes или no). Номинативная переменная;
#### 22. internet - наличие интернета дома (yes или no). Номинативная переменная;
#### 23. romantic - в романтических отношениях (yes или no). Номинативная переменная;
#### 24. famrel - семейные отношения (от 1 - очень плохо до 5 - очень хорошо). Количественная переменная;
#### 25. freetime - свободное время после школы (от 1 - очень мало до 5 - очень мого). Количественная переменная;
#### 26. goout - проведение времени с друзьями (от 1 - очень мало до 5 - очень много). Количественная переменная;
#### 27. health - текущее состояние здоровья (от 1 - очень плохо до 5 - очень хорошо). Количественная переменная;
#### 28. absences - количество пропущенных занятий. Количественная переменная;
#### 29. score - баллы по госэкзамену по математике. Количественная переменная.

## 3. Этапы работы над проектом
#### 1. Рассмотрение и первичная обработка данных; 
#### 2. Проверка данных на пустые значения;
#### 3. Работа с количественными переменными:
#### 3.1 Построение гистограмм распределения признака для количественных переменных, устранение выбросов;
#### 3.2 Заполнение пропусков для количественных переменных;
#### 3.3 Корреляционный анализ количественных переменных и отбор некоррелирующих переменных;
#### 3.4 Промежуточные выводы работы с количественными переменными;
#### 4. Работа с номинативными переменными:
#### 4.1 Анализ номинативных переменных;
#### 4.2 Заполнение пропусков для номинативных переменных;
#### 4.3 Отбор номинативных переменных для будущей модели;
#### 5. Основные итоги работы над проектом.
